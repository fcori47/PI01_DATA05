```python
#Debido a que si bien la columna 'rating' hace referencia a temporadas pero estamos
# mostrando información de películas voy a decidir poner los datos como NaN

all_frames.rating = all_frames.rating.replace(['1 Season',np.nan])
all_frames.rating = all_frames.rating.replace(['2 Seasons',np.nan])
all_frames.rating = all_frames.rating.replace(['3 Seasons',np.nan])
all_frames.rating = all_frames.rating.replace(['4 Seasons',np.nan])
all_frames.rating = all_frames.rating.replace(['6 Seasons',np.nan])
all_frames.rating = all_frames.rating.replace(['12 Seasons',np.nan])
```

```python
#Al corroborar de que no hay más valores que hacen referencia a temporadas de series 
# me doy cuenta de que tambien hay valores que hacen referencia a minutos, por lo que voy a dejarlos como NaN.

all_frames.rating.value_counts()
```
